```python
# filtering based off of school_code:

department_summary = df_filtered.groupby('dept_code')['contains'].agg(
        total_types='count',
        matching_types='sum'
    )
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(department_summary)

department_summary.to_csv('department_summary.csv')
```

```python
department_summary['percentage_matching'] = (
    department_summary['matching_types'] / department_summary['total_types'].replace(0, float('nan'))
) * 100

department_summary.to_csv('department_summary.csv')

department_summary['total_types'] = department_summary['total_types'].astype(int)

print(department_summary[department_summary['percentage_matching'] > 40].shape[0])
print(department_summary.shape[0])

display(department_summary.sort_values(by="percentage_matching", ascending=True))
```
